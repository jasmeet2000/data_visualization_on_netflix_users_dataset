```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset into a Pandas DataFrame
df = pd.read_csv('netflix_userbase.csv')

# Create a figure with four subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Visualization 1: Age Distribution (Histogram)
axes[0].hist(df['Age'], bins = 10, color = 'skyblue')
axes[0].set_xlabel('Age', fontweight = 'bold')
axes[0].set_ylabel('Netflix Users', fontweight = 'bold')
axes[0].set_title('Distribution of Users by Age', fontweight = 'bold', fontsize = 16)

# Visualization 2: Netflix Usage Based on Gender (Bar Chart)
gender_counts = df['Gender'].value_counts()
gender_colors = ['pink', 'lightblue']
axes[1].bar(gender_counts.index, gender_counts.values, color = gender_colors)
axes[1].set_xlabel('Gender', fontweight='bold')
axes[1].set_ylabel('Netflix Users', fontweight='bold')
axes[1].set_title('Netflix Usage Based on Gender', fontweight='bold', fontsize = 16)

# Adjust the layout
plt.tight_layout()
plt.show()
```

```python
# Visualization 3: Geographical Distribution of Netflix Users (Bar Chart)

country_counts = df['Country'].value_counts()
country_colors = ['red', 'green', 'lightcoral', 'magenta', 'salmon', 'lightgreen', 'lightblue', 'coral', 'blue', 'yellow']
plt.barh(country_counts.index, country_counts.values, color = country_colors)
plt.xlabel('Netflix Users', fontweight = 'bold')
plt.ylabel('Country', fontweight = 'bold')
plt.title('Geographical Distribution of Netflix Users', fontweight = 'bold', fontsize = 16)

# Adjust the layout
plt.tight_layout()
plt.show()
```

# Subscription Type

### Subscription:

1. Distribution
2. Revenue

```python
import pandas as pd
import matplotlib.pyplot as plt


# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Define a custom color palette
custom_subscription = ['lightgreen', 'lightblue', 'magenta']

# Visualization 4: The Distribution of Subscription Type (Bar Chart)
subscription_counts = df['Subscription Type'].value_counts()
axes[0].bar(subscription_counts.index, subscription_counts.values, color = custom_subscription) 
axes[0].set_xlabel('Subscription Type', fontweight = 'bold')
axes[0].set_ylabel('Netflix Users',  fontweight = 'bold')
axes[0].set_title('Distribution of Subscription Types', fontweight = 'bold', fontsize = 16)


#  Visualization 5: The revenue distribution by subscription type (Box Plot)
subscription_order = df['Subscription Type'].unique()
custom_palette     = ['lightgreen','magenta', 'lightblue']
medians            = []                               

for i, sub_type in enumerate(subscription_order):
    subset_data = df[df['Subscription Type'] == sub_type]
    bp = axes[1].boxplot(subset_data['Monthly Revenue'], positions=[i], widths=0.6, patch_artist=True, 
                         boxprops=dict(facecolor=custom_palette[i % len(custom_palette)]), medianprops=dict(color='black'))
    medians.append(bp['medians'][0].get_ydata()[0])      

axes[1].set_xticks(range(len(subscription_order)))
axes[1].set_xticklabels(subscription_order)
axes[1].set_xlabel('Subscription Type', fontweight='bold')
axes[1].set_ylabel('Monthly Revenue', fontweight='bold')
axes[1].set_title('Revenue Distribution by \nSubscription Type', fontweight='bold', fontsize=16)

# Highlight the median line using a black line
for i, sub_type in enumerate(subscription_order):
    axes[1].plot(i, medians[i], color='black')

# Adjust the layout
plt.tight_layout()
plt.show()
```

# DEVICE

### Device:
1. Usage Distribution, and 
2. It's Revenue 

```python
import pandas as pd
import matplotlib.pyplot as plt


# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Define a custom color palette
device_counts  = df['Device'].value_counts()
custom_palette = ['lightblue', 'lightgreen', 'lightcoral', 'magenta']

# Visualization 6: Device Usage Distribution (Bar Chart) 
axes[0].barh(device_counts.index, device_counts.values, color = custom_palette)
axes[0].set_xlabel('Netflix Users', fontweight = 'bold')
axes[0].set_ylabel('Device Used', fontweight = 'bold')
axes[0].set_title('Device Usage Distribution', fontweight = 'bold', fontsize = 16)


# Visualization 7: The revenue distribution by device (Box Plot)
medians = []   
# Extract unique device types to define the order for the box plot
device_order = df['Device'].unique()
custom_palette = ['lightcoral', 'lightgreen', 'lightblue', 'magenta']
for i, device in enumerate(device_order):
    subset_data = df[df['Device'] == device]
    bp = axes[1].boxplot(subset_data['Monthly Revenue'], positions=[i], widths=0.6, patch_artist=True, 
                         boxprops=dict(facecolor=custom_palette[i % len(custom_palette)]), 
                         medianprops=dict(color='black'))         
    medians.append(bp['medians'][0].get_ydata()[0])               
    
axes[1].set_xticks(range(len(device_order)))
axes[1].set_xticklabels(device_order)
axes[1].set_xlabel('Device Used', fontweight='bold')
axes[1].set_ylabel('Monthly Revenue', fontweight='bold')
axes[1].set_title('Revenue Distribution by Device', fontweight='bold', fontsize=16)

# Highlight the median line using a black line
for i, device in enumerate(device_order):
    axes[1].plot(i, medians[i], color='black')

# Adjust the layout
plt.tight_layout()
plt.show()
```
